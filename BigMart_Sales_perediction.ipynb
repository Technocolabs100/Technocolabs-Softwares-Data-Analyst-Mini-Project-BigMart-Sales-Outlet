No Outliers Detected 

#### Item_outlet_Sales

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(1,1,1)
df.boxplot(column=['Item_Outlet_Sales'], ax=ax)
plt.show()
```

```python
Q1 = df['Item_Outlet_Sales'].quantile(0.25)
Q3 = df['Item_Outlet_Sales'].quantile(0.75)
IQR = Q3 - Q1
outliers = ((df['Item_Outlet_Sales'] < (Q1 - 1.5 * IQR)) | (df['Item_Outlet_Sales'] > (Q3 + 1.5 * IQR)))
num_outliers = len(df[outliers])
print("Number of outliers:", num_outliers)
```

#### Replace the outlier value using the IQR

```python
def remove_outlier(col):
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1
    lower_range = Q1 - 1.5 * IQR
    upper_range = Q3 + 1.5 * IQR
    return lower_range, upper_range
```

```python
lower_range, upper_range = remove_outlier(df['Item_Outlet_Sales'])
df = df[(df['Item_Outlet_Sales'] >= lower_range) & (df['Item_Outlet_Sales'] <= upper_range)]
```

## Data Normalization
